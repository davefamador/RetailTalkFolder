# Data Cleaning — IntentDataset.xlsx

This notebook performs data cleaning on `IntentDataset.xlsx`:
- Load the dataset
- Identify and remove duplicate rows
- Save the cleaned dataset

```python
import pandas as pd
import os
```

## 1. Load the Dataset

```python
DATASET_DIR = r"c:\Moi\Thesis\Code\RetailTalkFolder\shopping_queries_dataset"
FILE_PATH = os.path.join(DATASET_DIR, "IntentDataset.xlsx")

df = pd.read_excel(FILE_PATH, engine='openpyxl')
print(f"Original shape: {df.shape}")
df.head()
```

## 2. Inspect for Duplicates

```python
num_duplicates = df.duplicated().sum()
print(f"Number of duplicate rows: {num_duplicates}")
```

```python
if num_duplicates > 0:
    print("Duplicate rows:")
    display(df[df.duplicated(keep=False)].sort_values(by=df.columns.tolist()))
else:
    print("No duplicate rows found.")
```

## 3. Remove Duplicates

```python
df_cleaned = df.drop_duplicates()
rows_removed = len(df) - len(df_cleaned)
print(f"Rows removed: {rows_removed}")
print(f"Cleaned shape: {df_cleaned.shape}")
```

## 4. Save Cleaned Dataset

```python
OUTPUT_PATH = os.path.join(DATASET_DIR, "IntentDataset_cleaned.xlsx")

df_cleaned.to_excel(OUTPUT_PATH, index=False, engine='openpyxl')
print(f"Cleaned dataset saved to: {OUTPUT_PATH}")
```

```python
# Quick verification
df_verify = pd.read_excel(OUTPUT_PATH, engine='openpyxl')
print(f"Verified shape: {df_verify.shape}")
print(f"Remaining duplicates: {df_verify.duplicated().sum()}")
```
